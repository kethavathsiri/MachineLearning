## Handling Missing Values

```python
#Replace the missing values in Embarked column by 'S'
df=pd.read_csv("train_titanic.csv")
df["Embarked"]=df["Embarked"].fillna('S')
print(df["Embarked"].isnull().sum())
```

```python
#Impute the missing values in Fare column with median value
df=pd.read_csv("train_titanic.csv")
df["Fare"]=df["Fare"].fillna(df["Fare"].median())
print(df["Fare"].isnull().sum())
```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data

import pandas as pd
import numpy as np

# Load dataset

df = pd.read_csv("train_titanic.csv")

# Get mean and standard deviation of the existing Age values
mean_age = df["Age"].mean()
std_age = df["Age"].std()

# Count missing values in Age
num_missing = df["Age"].isnull().sum()

# Generate random ages with same mean and standard deviation
random_ages = np.random.normal(mean_age, std_age, num_missing)

# Replace missing Age values with generated random ages
df.loc[df["Age"].isnull(), "Age"] = random_ages

# Verify if missing values are replaced
print(df["Age"].isnull().sum())  # Should print 0
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load dataset
df = pd.read_csv("train_titanic.csv")

# Encode 'Sex' using LabelEncoder
df['Sex'] = LabelEncoder().fit_transform(df['Sex'])

# One-Hot Encode 'Embarked' (excluding the first category to avoid dummy variable trap)
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True).replace({True:1,False:0})

# Display the relevant part of the processed DataFrame
print(df[['Sex', 'Embarked_Q', 'Embarked_S']].head())
```

```python
#Do a pairplot of all features

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("train_titanic.csv")

# Encode categorical features before plotting
from sklearn.preprocessing import LabelEncoder

# Copy data and encode 'Sex'
df_encoded = df.copy()
df_encoded["Sex"] = LabelEncoder().fit_transform(df_encoded["Sex"])

# One-Hot Encoding for 'Embarked'
df_encoded = pd.get_dummies(df_encoded, columns=["Embarked"], drop_first=True)

# Drop non-numeric columns (like Name, Ticket, Cabin, PassengerId)
df_encoded = df_encoded.select_dtypes(include=["number"])

# Plot pairplot
sns.pairplot(df_encoded, hue="Survived", diag_kind="kde")
plt.show()
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
from sklearn.model_selection import train_test_split

# Define features (X) and target variable (y)
X = df.drop(columns=["Survived"])  # Features
y = df["Survived"]  # Target variable

# Drop non-numeric columns
X = X.drop(columns=["Name", "Ticket", "Cabin"], errors="ignore")

# Encode categorical variables
X["Sex"] = LabelEncoder().fit_transform(X["Sex"])
X = pd.get_dummies(X, columns=["Embarked"], drop_first=True)  # One-Hot Encoding

# Split data (90% train, 10% validation)
X_train, X_validation, y_train, y_validation = train_test_split(
    X, y, test_size=0.1, random_state=42, stratify=y
)
```

```python
X_train.shape
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library
# Import necessary library

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder



# Handle missing values using SimpleImputer
imputer = SimpleImputer(strategy="median")
X_train = pd.DataFrame(imputer.fit_transform(X_train), columns=X.columns)
X_validation = pd.DataFrame(imputer.transform(X_validation), columns=X.columns)

# Train the Decision Tree model
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)


```

```python
#Print the tree
# Importing required packages for visualization
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Set figure size for better readability
plt.figure(figsize=(15, 8))

# Plot the decision tree
plot_tree(clf, feature_names=X.columns, class_names=["Not Survived", "Survived"], filled=True, rounded=True)

# Show the tree plot
plt.show()
```

```python
# plotting tree with max_depth=3
plt.figure(figsize=(50, 30))

# Plot the tree with max_depth=3
plot_tree(clf, feature_names=X.columns, class_names=["Not Survived", "Survived"], 
          filled=True, rounded=True, max_depth=3)

# Show the tree plot
plt.show()
```

```python
# Check the evaluation metrics of our default model on validation set - Confusion Matrix, Accuracy, Precision, Recall, F1 score

# Importing classification report and confusion matrix from sklearn metrics

# Making predictions

# Printing classification report
# Confusion Matrix
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Make Predictions
y_pred = clf.predict(X_validation)

# Confusion Matrix
cm = confusion_matrix(y_validation, y_pred)
print("Confusion Matrix:\n", cm)

# Plot Confusion Matrix for better visualization
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=["Not Survived", "Survived"], 
            yticklabels=["Not Survived", "Survived"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

plt.show()

# Accuracy
accuracy = accuracy_score(y_validation, y_pred)

# Classification Report (Precision, Recall, F1-score)
print("Classification Report:\n", classification_report(y_validation, y_pred))

print(f"Accuracy: {accuracy:.2f}")


```
