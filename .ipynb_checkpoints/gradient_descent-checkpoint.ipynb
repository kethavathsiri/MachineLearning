```python
import numpy as np

def gradient_descent(x, y):
    m_curr = b_curr = 0  # Initialize slope and intercept
    iterations = 100
    n = len(x)
    learning_rate = 0.01

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr  # Fixing the equation
        cost = (1/n) * sum((y - y_predicted) ** 2)  # Cost function
        
        md = -(2/n) * sum(x * (y - y_predicted))  # Partial derivative wrt m
        bd = -(2/n) * sum(y - y_predicted)  # Partial derivative wrt b
        
        m_curr = m_curr - learning_rate * md  # Updating m
        b_curr = b_curr - learning_rate * bd  # Updating b
        
        print("m={}, b={}, cost={}, iteration={}".format(m_curr, b_curr, cost, i))  # Correct format

# Define training data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Call the function
gradient_descent(x, y)
```
