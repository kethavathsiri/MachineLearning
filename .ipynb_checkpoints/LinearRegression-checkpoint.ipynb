```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/home/user/MachineLearning/py-master/py-master/ML/1_linear_reg/homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('Area(sq ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='*')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)#training the LR(Linear Regression model)
```

```python
#import pandas as pd
#reg.predict([[3300]])
reg.predict(pd.DataFrame([[3300]], columns=['area']))
```
