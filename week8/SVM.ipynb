```python
#19 prepare x and y
df=pd.read_csv("Iris.csv")
X=df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
y=df["Species"]
```

```python
#20 Split into train and test data-70% trian random state=1
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1, stratify=y
)
```

```python
#21 Look at the documentation of SVC class that you will use to create the model
from sklearn.svm import SVC
model=SVC(C=0.1,kernel='rbf',gamma='scale', random_state=1)
model.fit(X_train, y_train)
```

```python
#22 Model building
from sklearn.metrics import accuracy_score, classification_report
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test, y_pred)
report=classification_report(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(report)
```

```python
#30 Use grid search CV to find optimal value of c-sel decision tree lab questions
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
import pandas as pd

# Load dataset
df = pd.read_csv('Iris.csv')

# Prepare features and labels
X = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
y = df["Species"]

# Define parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf']  # Optional: trying different kernels too
}

# Initialize SVC
svc = SVC()

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, cv=5)

# Fit on full data
grid_search.fit(X, y)

# Print the best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python
#31 Try with C=[0.1,1,10,100,1000]
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
import pandas as pd

# Load dataset
df = pd.read_csv('Iris.csv')

# Prepare features and labels
X = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
y = df["Species"]

# Define parameter grid with extended C values
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'kernel': ['linear', 'rbf']  # Optional: try different kernels
}

# Initialize the SVC model
svc = SVC()

# Grid search with 5-fold cross-validation
grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, cv=5)

# Fit the model
grid_search.fit(X, y)

# Print best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Accuracy:", grid_search.best_score_)
```

```python
#35 PLot of c verrus train and test scores
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Define parameter grid
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'kernel': ['linear']
}

# Create the model
svc = SVC()

# Run grid search with return_train_score=True
grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, cv=5, return_train_score=True)
grid_search.fit(X, y)

# Convert results to DataFrame
results = pd.DataFrame(grid_search.cv_results_)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(results['param_C'], results['mean_train_score'], marker='o', label='Train Score')
plt.plot(results['param_C'], results['mean_test_score'], marker='s', label='Test Score')
plt.xscale('log')  # Log scale for better visualization
plt.xlabel('C (Regularization Parameter)')
plt.ylabel('Accuracy')
plt.title('Train vs Test Accuracy for different C values')
plt.legend()
plt.grid(True)
plt.show()
```
