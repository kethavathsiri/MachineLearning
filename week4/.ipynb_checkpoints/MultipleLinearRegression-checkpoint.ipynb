## Housing Case Study
Problem Statement:

Consider a real estate company that has a dataset containing the prices of properties in the Delhi region. It wishes to use the data to optimise the sale prices of the properties based on important factors such as area, bedrooms, parking, etc.

Essentially, the company wants —

To identify the variables affecting house prices, e.g. area, number of rooms, bathrooms, etc.

To create a linear model that quantitatively relates house prices with variables such as number of rooms, area, number of bathrooms, etc.

To know the accuracy of the model, i.e. how well these variables can predict house prices.

```python
#7 Encoding categorical data in X
df=pd.read_csv("../week4/Housing.csv")
cols = ["mainroad", "guestroom", "basement", "hotwaterheating", "airconditioning", "prefarea"]
for c in cols:
    df[c] = df[c].map({"yes": 1, "no": 0})

dummies = pd.get_dummies(df["furnishingstatus"]).replace({True:1,False:0})
merge = pd.concat([df.drop("furnishingstatus", axis=1), dummies], axis=1)
dummies
merge
```

```python
#8  Avoiding the Dummy Variable Trap

dummies = pd.get_dummies(df["furnishingstatus"], drop_first=True).replace({True:1,False:0})
dummies
```

```python
#9 Apply feature scaling on numerical variables
from sklearn.preprocessing import StandardScaler

# Separate X and y
X = merge.drop("price", axis=1)
y = merge["price"]

# List of numerical columns to scale
numerical_cols = ["area", "bedrooms", "bathrooms", "stories", "parking"]

# Apply StandardScaler
scaler = StandardScaler()
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

# Optional: check result
print(X.head())
```

```python
#10 Split Data into Training and Testing Sets (70%-Train 30%-Test) - X_train, y_train, X_test, y_test
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
print("Training set:", X_train.shape, y_train.shape)
print("Testing set:", X_test.shape, y_test.shape)
```

```python
#11 Fitting Multiple Linear Regression to the Training
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#12 Predict on the train set and calculate the error = y_pred - y_train
y_predict_train=model.predict(X_train)
error=y_predict_train-y_train
print("Prediction Error (y_pred - y_train):")
print(error.head())
```

```python
#13  Residual plot - plot y_pred in x axis and errors in y axis
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure y_pred is a pandas Series with same index as y_train
y_pred_train = pd.Series(model.predict(X_train), index=y_train.index)
residuals = y_pred_train - y_train

# Plotting residuals
plt.figure(figsize=(8, 5))
sns.scatterplot(x=y_pred_train, y=residuals)
plt.axhline(0, color='red', linestyle='--')  # Line at y=0 for reference
plt.xlabel("Predicted values (y_pred)")
plt.ylabel("Residuals (y_pred - y_actual)")
plt.title("Residual Plot (Training Set)")
plt.tight_layout()
plt.show()
```

```python
#14  Predict on the test set
y_pred_test = model.predict(X_test)

# Optional: Convert to pandas Series for easier handling
import pandas as pd
y_pred_test = pd.Series(y_pred_test, index=y_test.index)

# View first few predictions
print("Predicted values on test set:")
print(y_pred_test.head())
```

```python
#15 Residual plot - plot y_pred in x axis and errors in y axis
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Make sure y_pred_test is a pandas Series aligned with y_test
y_pred_test = pd.Series(y_pred_test, index=y_test.index)

# Calculate residuals
residuals_test = y_pred_test - y_test

# Plotting residuals
plt.figure(figsize=(8, 5))
sns.scatterplot(x=y_pred_test, y=residuals_test)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicted values (y_pred_test)")
plt.ylabel("Residuals (y_pred - y_actual)")
plt.title("Residual Plot (Test Set)")
plt.tight_layout()
plt.show()
```

```python
#16 Print Mean Squared Error and R Squared Value
from sklearn.metrics import mean_squared_error, r2_score

# Calculate MSE
mse = mean_squared_error(y_test, y_pred_test)

# Calculate R² score
r2 = r2_score(y_test, y_pred_test)

# Print results
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R² Score: {r2:.4f}")
```

```python
#17 Check  Adjusted R Squared Value(by selecting different no of input variables insted of all)
# Number of observations and features
n = X_test.shape[0]   # rows in test set
k = X_test.shape[1]   # number of input variables

# Compute adjusted R²
adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - k - 1))

print(f"Adjusted R² Score: {adjusted_r2:.4f}")
```
