```python
#5
from sklearn.cluster import KMeans
import pandas as pd
df = pd.read_csv("Iris.csv")
sepal_data = df[['SepalLengthCm', 'SepalWidthCm']].copy()
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0, n_init=10)
    kmeans.fit(sepal_data)
    wcss.append(kmeans.inertia_)
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
kmeans.fit(sepal_data)
sepal_data['Cluster'] = kmeans.labels_
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load data and select features
df = pd.read_csv("Iris.csv")
X = df.iloc[:, 1:5]  # Selecting Sepal & Petal features

# Elbow method to find optimal k
k_values = []
wcss_values = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, max_iter=100, n_init=10, random_state=0)
    kmeans.fit(X)
    k_values.append(k)
    wcss_values.append(kmeans.inertia_)
    print(f'k = {k}, WCSS = {kmeans.inertia_}')


```

```python
#8. Plotting the wcss
plt.plot(k_values, wcss_values, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

```python
#9
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
#10
print("Cluster labels for each point:")
print(y_kmeans)
```

```python
#11
plt.scatter(X.iloc[y_kmeans == 0, 0], X.iloc[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X.iloc[y_kmeans == 1, 0], X.iloc[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(X.iloc[y_kmeans == 2, 0], X.iloc[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(X.iloc[y_kmeans == 3, 0], X.iloc[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(X.iloc[y_kmeans == 4, 0], X.iloc[y_kmeans == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 
            s=300, c='yellow', label='Centroids')

plt.title('Clusters of Iris Data')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()
plt.show()
```
